```python
import requests
import pandas as pd
import json
```

```python
url = "https://api.openaq.org/v2/cities?limit=10000&page=1&offset=0&sort=asc&country_id=US&order_by=city"

headers = {"accept": "application/json"}

response = requests.get(url, headers=headers)

print(response.text)
```

```python
data = response.json()["results"]

df = pd.json_normalize(data)
df.head()
```
